## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Total number of person using international plan
intl_user=df_telecom.loc[df_telecom['International plan']=='Yes'].State.count()
print("The total number of customer uses international plan are:\n",intl_user)
# Total number of person using voice mail plan
voicemail_user=df_telecom.loc[df_telecom['Voice mail plan']=='Yes'].State.count()
print("The total number of customer uses voice mail plan are:\n",voicemail_user)
# Total number of person using voice mail plan and international plan
both_user=df_telecom.loc[(df_telecom['Voice mail plan']=='Yes')&(df_telecom['International plan']=='Yes')].State.count()
print("The total number of customer uses both voice mail plan and international plan are:\n",both_user)
# Total number of person nither using voice mail plan and nor international plan
nboth_user=df_telecom.loc[(df_telecom['Voice mail plan']=='No')&(df_telecom['International plan']=='No')].State.count()
print("The total number of customer uses niether voice mail plan and nor international plan are:\n",nboth_user)
```

```python
# Write your code to make your dataset analysis ready.
df_telecom_churn=df_telecom.loc[df_telecom['Churn']==True]
df_telecom_churn
```

```python
print(f"Total number of people churned are:\n{df_telecom.loc[df_telecom['Churn']==True].Churn.count()}")
print(f"Percentage of people Churned are:\n{(483/3333)*100}")
print(f"Total number of people use international plan churned are:\n{df_telecom_churn.loc[df_telecom_churn['International plan']=='Yes'].Churn.count()}")
print(f"Total number of people use Voice mail plan churned are:\n{df_telecom_churn.loc[df_telecom_churn['Voice mail plan']=='Yes'].Churn.count()}")
print(f"Total number of people use both Voice mail plan and international plan churned are:\n{df_telecom_churn.loc[(df_telecom_churn['Voice mail plan']=='Yes')&(df_telecom_churn['International plan']=='Yes')].Churn.count()}")
print(f"Total number of people neither use Voice mail plan and nor international plan churned are:\n{df_telecom_churn.loc[(df_telecom_churn['Voice mail plan']=='No')&(df_telecom_churn['International plan']=='No')].Churn.count()}")
```

```python
#percentage of people churned
print(f"Total percentage of people churned are:\n{(483/3333)*100}")
print(f"Total percentage of people churned using international plan  are:\n{(137/323)*100}")
print(f"Total percentage of people churned using Voice mail plan  are:\n{(80/922)*100}")
print(f"Total percentage of people churned using no special plan  are:\n{(302/2180)*100}")
```

```python
# we will find number of people churned with respect to state
df_statechurn=df_telecom_churn.groupby('State').agg({'Churn':'count'}).reset_index().sort_values('Churn',ascending=False)
df_statechurn
```

```python
# we will find number of people churned with respect to Area code
df_areachurn=df_telecom_churn.groupby('Area code').agg({'Churn':'count'}).reset_index().sort_values('Churn',ascending=False)
df_areachurn
```

```python
# we will find number of people churned uses in day
df_telecom.groupby('Churn').agg({'Total day minutes':'sum','Total day calls':'sum','Total day charge':'sum'}).reset_index()
```

```python
# we will find number of people churned uses in evening
df_telecom.groupby('Churn').agg({'Total eve minutes':'sum','Total eve calls':'sum','Total eve charge':'sum'}).reset_index()
```

```python
# we will find number of people churned uses in night
df_telecom.groupby('Churn').agg({'Total night minutes':'sum','Total night calls':'sum','Total night charge':'sum'}).reset_index()
```

```python
# we will find number of people churned uses international calls
df_telecom.groupby('Churn').agg({'Total intl minutes':'sum','Total intl calls':'sum','Total intl charge':'sum'}).reset_index()
```

```python
df_servicecall=df_telecom_churn.groupby('Area code').agg({'Customer service calls':'sum','Churn':'count'}).reset_index()
df_servicecall
```
